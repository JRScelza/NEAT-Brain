```python
import os
import numpy as np
import pandas as pd
import ipywidgets as widgets
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
data_directory = os.path.dirname(os.getcwd())+'/data'
df = pd.read_csv(data_directory + "/cloutier_2009_data.csv")
df.head(10)
```

```python
reaxn = widgets.Dropdown(
    options=list(zip(df.columns,df.index))[1:],
    value=2,
    description='Number:',
)
reaxn
```

```python
print(' ')
print(reaxn.label)

f = plt.figure()
plt.plot(range(len(df[reaxn.label])), df[reaxn.label])
plt.show()
```
