```python
df = pd.read_csv("/Users/jeromescelza/Library/CloudStorage/OneDrive-JohnsHopkins/Game Theory/NEAT-Brain/data/cloutier_2009_data.csv")
df.head()
```

```python
%matplotlib inline

fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(15, 5))
plt.subplots_adjust(hspace=0.5)
count = 0

for i in df.columns:
    
    if count <5:
        axs[count].plot(df[i],c=np.random.rand(3,), label=i)
        axs[count].legend()
        count +=1
        
    else:
        count = 0
        fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(15, 5))
        plt.subplots_adjust(hspace=0.5)
```
